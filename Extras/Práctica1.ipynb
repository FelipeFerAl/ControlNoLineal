```python
#@title Ejemplo Manual 1.4 del Libro

import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 360, 60)
t2 = np.linspace(0, 360, 120)

x = np.cos(np.deg2rad(t))
y = np.sin(np.deg2rad(t))

x2 = 2*np.cos(np.deg2rad(t2))
y2 = 2*np.sin(np.deg2rad(t2))

plt.figure(figsize=(10, 10))
plt.plot(x, y, 'o')
plt.plot(x2, y2, '*')
plt.show()
```

```python
#@title Ejemplo 1.4 del libro completo
def linear_oscillator(x, y):
  dx = y
  dy = -x
  return dx, dy

min, max = -2*np.pi, 2*np.pi
low, high = -2, 2

# Grid
theta = np.linspace(min, max, 600)
omega = np.linspace(low, high, 400)
THETA, OMEGA = np.meshgrid(theta, omega)

# Vector field
dTHETA, dOMEGA = linear_oscillator(THETA, OMEGA)
```

```python
#@title Diagrama
plt.figure(figsize=(12, 5))

plt.streamplot(
  THETA, OMEGA,
  dTHETA, dOMEGA,
  density=1.4,
  linewidth=0.8,
  arrowsize=1
)

# Axes
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

# Ticks at multiples of Ï€
ticks = [-3*np.pi, -np.pi, np.pi, 3*np.pi]
labels = [r"$-3\pi$", r"$-\pi$", r"$\pi$", r"$3\pi$"]

plt.xticks(ticks, labels)

plt.xlim(min, max)
plt.ylim(low, high)

for k in range(-3, 4):
  plt.plot(2*k*np.pi, 0, 'ko', markersize=4)

plt.grid(False)
plt.show()
```
